```python
# imports
import requests
import time
from bs4 import BeautifulSoup
import os

to = .05
```

```python
def get_sub_urls(url, first=0, last=0, fatcheck=False):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    sub_urls = soup.findAll("a")
    sub_urls = sub_urls[first:len(sub_urls)+last]
    if fatcheck:
        sub_urls_fatchecked = []
        for elem in sub_urls:
            name = next(elem.children).name
            elem = elem["href"]
            full_path = elem if elem[:4] == "http" else os.path.join(os.path.dirname(url), elem)
            new_elem = (full_path, name == "strong")
            sub_urls_fatchecked.append(new_elem)
        sub_urls = [elem for elem in sub_urls_fatchecked if elem[0][-3:]=="htm" or elem[0][-4:]=="html"]
    else:
        sub_urls = [elem["href"] for elem in sub_urls]
        sub_urls = [elem if elem[:4] == "http" else os.path.join(os.path.dirname(url), elem) for elem in sub_urls]
        sub_urls = [elem for elem in sub_urls if elem[-3:]=="htm" or elem[-4:]=="html"]
    return sub_urls
```

```python
def download_page(url, destination):
    response = requests.get(url)
    file_name = os.path.basename(url)
    with open(os.path.join(destination, file_name), "w") as file:
        file.write(response.text)
```

```python
entry_url = "http://www.mlwerke.de/le/default.htm"
author = "lenin"
baende_urls = get_sub_urls(entry_url, 3, -3, fatcheck=False)
baende_urls
```

```python
# for single texts
already_scraped = []
for text_url in baende_urls:
    if text_url in already_scraped:
        continue
    download_page(text_url, author)
    already_scraped.append(text_url)
    time.sleep(to)
```

```python
import re

# for werkausgaben
already_scraped = []

for band_url in baende_urls:
    dir_name = os.path.basename(os.path.dirname(band_url))
    #dir_name = band_url[45:-7]
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)
    text_urls = get_sub_urls(band_url, 1, -1)
    for text_url in text_urls:
        if text_url in already_scraped:
            continue
        print(f"Downloading from {text_url}")
        download_page(text_url, dir_name)
        already_scraped.append(text_url)
        time.sleep(to)
    
    time.sleep(to)
```

```python
# for weird mixtures -> fatcheck must have been True
already_scraped = []
for uso in baende_urls:
    if uso[1]:
        dir_name = os.path.basename(os.path.dirname(uso[0]))
        dir_name = os.path.join(author, dir_name)
        if not os.path.exists(dir_name):
            os.mkdir(dir_name)
        text_urls = get_sub_urls(uso[0], 1, -1)
        for text_url in text_urls:
            if text_url in already_scraped:
                continue
            download_page(text_url, dir_name)
            already_scraped.append(text_url)
            time.sleep(to)  
        time.sleep(to)
        
    else:
        if uso[0] in already_scraped:
            continue
        download_page(uso[0], author)
        already_scraped.append(uso)
        time.sleep(to)
```
